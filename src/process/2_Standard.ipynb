# Address

```python
#Lọc dữ liệu trùng p2 + Matching

#Tien xu li matching
#Bước 1: Xây dựng bot lấy dữ liệu từ google map về kinh độ, vĩ độ theo address
#Ở file python khác


#Bước 2: Xử lí address hỗ trợ cho string matching
strlist=['street','district','ward','phuong','duong','quan','thanh pho','tp','ha noi','da nang','ho chi minh','tp.',',','hcmc','city','viet nam','100000','50000','51000','55000','700.000','700000','701000','70000','000000','70000','00000','70 000','710200','0000','000',' ']
#ebooking
ebooking['new_add']=ebooking['address'].str.lower()
ebooking['new_add']=[unidecode.unidecode(y) for y in ebooking['new_add'].fillna('NA') ]
ebooking['new_add'].replace(strlist,'', inplace=True, regex=True)
ebooking['new_add']=ebooking['new_add'].fillna('NA')

#traveloka
traveloka['new_add']=traveloka['address'].str.lower()
traveloka['new_add']=[unidecode.unidecode(y) for y in traveloka['new_add'].fillna('NA') ]
traveloka['new_add'].replace(strlist,'', inplace=True, regex=True)
traveloka['new_add']=traveloka['new_add'].fillna('NA')

#booking
booking['new_add']=booking['address'].str.lower()
booking['new_add']=[unidecode.unidecode(y) for y in booking['new_add'].fillna('NA') ]
booking['new_add'].replace(strlist,'', inplace=True, regex=True)
booking['new_add']=booking['new_add'].fillna('NA')


#agoda
agoda['new_add']=agoda['address'].str.lower()
agoda['new_add']=[unidecode.unidecode(y) for y in agoda['new_add'].fillna('NA') ]
agoda['new_add'].replace(strlist,'', inplace=True, regex=True)
agoda['new_add']=agoda['new_add'].fillna('NA')
```
