```python
def process_img(s):
    res = s.strip('][').split(', ')
    res1 = []
    for i in res:
        res1.append(i.replace("'", ''))
    return res1
```

```python
re = process_img(data.img_url[0])
```

```python
re[0]
```

```python
data['img1'] = data.img_url.apply(lambda x : process_img(x))
```

```python
data['ne'] = data.img1.apply(lambda x : len(x))
```

```python

data.drop(data[data['ne'] == 1].index, inplace = True)
```

```python
data['img_url1'] = data.img1.apply(lambda x : x[0])
data['img_url2'] = data.img1.apply(lambda x : x[1])
data['img_url3'] = data.img1.apply(lambda x : x[2])
```

```python
data['des1'] = data['des1'].apply(lambda x: x.replace('\\n', ' '))
```

```python
data['nearby_places'] = data.nearby_places.apply(lambda x: process_img(x))
data['facilities'] = data.facilities.apply(lambda x: process_img(x))

data['nearby_places'] = data['nearby_places'].apply(lambda x: ', '.join([str(i) for i in x]))
data['facilities'] = data['facilities'].apply(lambda x: ', '.join([str(i) for i in x]))
```

```python
def convert_city(city):
    if city == 'Hồ Chí Minh':
        return 3
    elif city == 'Hà Nội':
        return 1
    elif city == 'Đà Nẵng':
        return 2
    elif city == 'Đà Lạt':
        return 4
    elif city == 'Nha Trang':
        return 5
    elif city == 'Phú Quốc':
        return 6
    elif city == 'Quảng Ninh':
        return 7
    elif city == 'Vũng Tàu':
        return 8
```

```python
final['city_id'] = final.city.apply(lambda x: convert_city(x))
```
